```python
import pandas as pd
import pandas_profiling as pdp
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("data/train_data.csv")
```

```python
df.info()
```

```python
F, T = df["ratings_disabled"].value_counts().values

plt.title("ratings_disabled", fontsize=25)
plt.bar(["False", "True"], [F, T], width=0.2)
for x, y in zip(range(2), [F, T]):
    plt.text(x, y, y, ha='center', va='bottom', fontsize=20)
```

```python
F, T = df["comments_disabled"].value_counts().values

plt.title("comments_disabled", fontsize=25)
plt.bar(["False", "True"], [F, T], width=0.2)
for x, y in zip(range(2), [F, T]):
    plt.text(x, y, y, ha='center', va='bottom', fontsize=20)
```
